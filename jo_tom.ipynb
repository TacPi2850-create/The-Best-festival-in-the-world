```python
convert_GBP = 1.16 # Sterlina britannica
convert_DKK = 0.13 # Corona danese
convert_AUD = 0.61 # Dollaro australiano
convert_CHF = 1.04 # Franco svizzero
convert_USD = 0.91 # Dollaro americano
convert_PLN = 0.22 # Zloty polacco
convert_MXN = 0.05 # Peso messicano
convert_SEK = 0.09 # Corona svedese
convert_CAD = 0.67 # Dollaro canadese
convert_NOK = 0.09 # Corona norvegese
convert_JPY = 0.0065 # Yen giapponese
convert_CLP = 0.0011 # Peso cileno
convert_ARS = 0.003 # Peso argentino
```

```python
# Conversione Economic Impact in Euro
def converti_euro(value):
    # Dizionario con i tassi di conversione
    conversion_rates = {
        '£': convert_GBP,
        '€': 1,  # Euro, nessuna conversione necessaria
        'USD': convert_USD,
        'MXN': convert_MXN,
        'AU$': convert_AUD,
        'ARS': convert_ARS,
        'CHF': convert_CHF,
        'DKK': convert_DKK,
        'SEK': convert_SEK,
        'CLP': convert_CLP,
        'JPY': convert_JPY,
        'PLN': convert_PLN,
        'NOK': convert_NOK,
        'CAD': convert_CAD,
    }
    
    for currency, rate in conversion_rates.items():
        if currency in value:
            numeric_value = float(
                value.replace(currency, '').replace(' million', '').replace(' billion', '').replace(' ', ''))
            if 'million' in value:
                return '{:.0f}'.format(numeric_value * rate * 1e6)
            elif 'billion' in value:
                return '{:.0f}'.format(numeric_value * rate * 1e9)
    
    return value  # Se non corrisponde a nessuna delle valute, restituisce il valore originale
```

```python
# Update colonna Economic Impact
df['Economic Impact'] = df['Economic Impact'].apply(converti_euro)
```
